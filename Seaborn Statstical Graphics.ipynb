###### An Introduction to Seaborn

Seaborn is a python library that is used for making statistical graphics.

It builds on top of <a href = "">matplotlib library</a> and and integrate closely with <a href="">pandas</a> data structure.

Seaborn helps you explore and understand your data.

```python
# importing seaborn

import seaborn as sns
```

<a>How to install seaborn?</a>

```python
# seaborn.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)

# apply the default theme
sns.set_theme()
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```
